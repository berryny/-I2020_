# Exercise 1

Create a function that takes a url, creates a `GET` request, executes it and then returns the content of the webpage.

````python
import requests

def url_content(url):
    x = requests.get(url)
    if x.status_code == 200:
        print('Content\n', x.content)
        print('Text\n', x.text)
    else:
        print('An error has occurred.')
        print(x.status_code)
    
url_content('https://www.theknowledgehouse.org/')
````

Show that we're able to use this class to create instances of a bot that will scrape a website for data.

````python
import requests
import urllib3

# Create a class called Bot that stores attributes url
class Bot:
    def __init__(self,url):
        self.url = url

#     and the methods from the previous exercises
    def webpage_tags(self,html_tag):
        self.html_tag = html_tag
        req = requests.get(self.url)
        soup = bs4.BeautifulSoup(req.text)
        return soup.findAll(html_tag)
    
#     Alter the methods to store attributes such as status_code, headers and content when the request is executed.
    def request_content(self,html_tag):
        req = requests.get("http://www.google.com")
        headers = {}
        status = req.status_code
        content = req.content

        if status == 200:
            for k,v in req.headers.items():
                headers[k] = v
        
        return status, headers, content
        
````

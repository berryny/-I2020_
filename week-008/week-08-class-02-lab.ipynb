# Exercise 2

Create a function that will take a list of web pages and return a list of all the anchor tags on those pages.

```python
# a list of web pages
# web_page_list = ['https://pypi.org/project/beautifulsoup4/', 'https://github.com/berryny', 'https://linkedin.com']
web_page_list = ['https://linkedin.com']

# return a list of all the anchor tags on those pages.
def webpages(pages):
    print('web_pages',pages)
    for url in pages:
        req = requests.get(url)
        if req.status_code == 200:
            html_text = req.text
            soup = BeautifulSoup(html_text)
            return soup.findAll('a')
```

```python
webpages(web_page_list)
```

# Exercise 3:

Create a function that accesses a web page and returns data from that page.

```python
# a function that accesses a web page
payload = dict(keyData='valueData')

def access_web_page(url):
#     returns data from that page.
    req = requests.get(url)
    if req.status_code == 405:
        rpost = requests.post(url, data=payload)
        print(rpost.text)
    else:
        print(req.text)
```

```python
access_web_page('https://httpbin.org/post')
```

```python
access_web_page('https://linkedin.com/')
```

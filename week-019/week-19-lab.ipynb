**Part 3**

Making observations of what we have.

**a.** What flight was the longest? Where did it leave from and where did it land?

You might need to look up airport codes.

````python
df["distance"].max()
````

````python
df.groupby("distance")[["origin", "dest"]].max()
````

````python
df.groupby("distance")[["origin", "dest"]].max().sort_values("distance", ascending=False).iloc[0]
#  integer-location [iloc]
````

**b.** What flight was the shortest?

````python
df.groupby("distance")[["origin", "dest"]].min().sort_values("distance").iloc[0]
#  integer-location [iloc]
````

**c.** What airline flied the most during this period?

hint: `.size()`

- [.value_counts()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.value_counts.html?highlight=value_count#pandas-dataframe-value-counts)
- [.to_frame()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.to_frame.html?highlight=to_frame#pandas-series-to-frame)

````python
df["airline_name"].value_counts().to_frame().iloc[0]
````

**d.** Which destination was most flown to?

````python
df["dest"].value_counts().to_frame().iloc[0]
````

**e.** Which location was flown from the most?

````python
df["origin"].value_counts().to_frame().iloc[0]
````

**f.** Which flight had the longest departure delay?

````python
df.groupby("dep_delay", as_index=True)[["origin", "dest", "airline_name", "flight"]].min().sort_values("dep_delay", ascending=False).iloc[0]
````

````python
df[df["dep_delay"] == df["dep_delay"].max()][["flight","dep_delay"]]
````

**g.** Which flight had the longest arrival delay? Should match the previous.

````python
df.groupby("arr_delay")[["origin", "dest", "airline_name", "flight"]].min().sort_values("arr_delay", ascending=False).iloc[0]
````

**h.** Which airline had the longest arrival delays on average?

````python
df.groupby(["airline_name"])["arr_delay"].mean().sort_values(ascending=False).to_frame().iloc[0]
````

**i.** Which airline had the longest departure delays on average?

````python
df.groupby(["airline_name"])["dep_delay"].mean().sort_values(ascending=False).to_frame().iloc[0]
````
